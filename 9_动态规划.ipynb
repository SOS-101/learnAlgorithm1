# 小结
需要在给定约束条件下优化某种指标时，动态规划很有用。❑ 问题可分解为离散子问题时，可使用动态规划来解决。❑ 每种动态规划解决方案都涉及网格。❑ 单元格中的值通常就是你要优化的值。❑ 每个单元格都是一个子问题，因此你需要考虑如何将问题分解为子问题。❑ 没有放之四海皆准的计算动态规划解决方案的公式。
